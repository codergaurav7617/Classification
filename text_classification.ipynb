# This is filtering each file to get pure data , first we remove all the stop words from the file and then filter out the \r the \n and all the stop words from the files to get pure words that work!

```python
from nltk.corpus import stopwords
import string
stop=stopwords.words('english')
punctuation=list(string.punctuation)
stop=stop+punctuation
stop
```

```python
def createArrayOfStrings(data):
    py=data.split(' ')
    #print(len(py))
    #print(py)
    arr=[]
    for x in range(len(py)):
        word=''
        word=py[x].replace('\n','')
        word=word.replace('\r','')
        word=word.replace('\t','')
        word=word.replace(',','')
        word=word.replace('(','')
        word=word.replace(').','')
        word=word.replace(')','')
        word=word.replace('.','')
        word=word.replace('"','')
        word=word.replace('!','')
        word=word.replace('>','')
        word=word.replace('<','')
        word=word.replace(':','')
        word=word.replace('|','')
        word=word.replace('[','')
        word=word.replace(']','')
        word=word.replace('@','')
        word=word.replace('?','')
        word=word.replace('--','')
        word=word.lower()
        if word=='':
            continue
        elif word in stop:
            continue
        else:
            arr.append(word)
    return arr
def filterLabels(labels):
    for x in range(len(labels)):
        labels[x]=labels[x].replace('20_newsgroups/','')
    return labels
```

```python
path="20_newsgroups/"
data=[]
labels=[]
for itemName in os.listdir(path):
    itemName = os.path.join(path, itemName)
    print (itemName)
    for myPath in os.listdir(itemName):
        myPath=os.path.join(itemName,myPath)
        file_readed = open(myPath,encoding = "ISO-8859-1")
        arr=createArrayOfStrings(file_readed.read())
        data.append(arr)
        labels.append(itemName)
```

```python
# data is now a 2d array, and now we are going to create a dictionary for every word in each data
labels=filterLabels(labels) 
dictonary={}
for x in data:
    for y in x:
        dictonary[y]=dictonary.get(y,0)+1#if y not there then place 1 and y in the dictionary as Value and Key resp...
print(len(dictonary.keys()))
# these are the features for the dataset and we are going to train the MNB on this dataset now!


# remove the elements which have the value 1 and value 2  in the dictionary 
for key in (dictonary.copy()).keys():
    if dictonary[key]==1 or dictonary[key]==2:
        del (dictonary[key])          
len(dictonary.keys())    
```

```python
# we are trying to convert dictonary in all_word List and using nltk module we are tying to find out most common 3000 word 
all_words = []
word=dictonary.keys()
for doc in word:
    for i in range(0,dictonary.get(doc,0)):
        if(doc.isdigit()):
            continue
        all_words.append(doc)   
```

```python
import nltk
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
features
```

```python
# From Scratch implementation of Naive bayes classifier
#we are trying to fit the mode by using fit function
def fit(X,labels):
    result={}
    sum1=0
    result["total_data"]=len(labels)
    classes=set(labels)
    for current_class in classes:
        result[current_class]={}
        num_feature=len(features)
        current_class_row=(labels==current_class)
        current_class_x=X[current_class_row]
        current_class_y=labels[current_class_row]
        result[current_class]["total_data"]=len(current_class_y)
        output=0
        for i in range(0,num_feature):
            word=word_dictonary[i]
            temp=(current_class_x[:,i]).sum()
            result[current_class][word]=temp
            output=output+temp
        result[current_class]["total_count"]=output    
    return result      
```

```python
# we are trying to predict the probablity of test data for that particular class
def probablity(dictonary,x,current_class):
    word_feature=list(dictonary.keys())
    output=np.log(dictonary[current_class]["total_data"])-np.log(dictonary["total_data"])
    num_fetures=len(dictonary[current_class].keys())
    for j in range(0,len(x)):
        xj=x[j]
        if xj==0:
            continue
        else:
            word=word_dictonary[j]
            count_current_class_with_value_xj=dictonary[current_class][word]+1
            count_current_class=dictonary[current_class]["total_count"]+len(word_feature)
            current_xj_probablity=np.log(count_current_class_with_value_xj)-np.log(count_current_class)
            output=output+current_xj_probablity
    return output       
```

```python
# we are here to predict the probablity of test data using all the classes and we are going to predict the class which are going to give the highest probablity 
def predictsinglepoint(dictonary,x):
    classes=dictonary.keys()
    best_p=-1000
    best_class='o'
    first_run=True
    for current_classes in classes:
        if(current_classes=="total_data"):
            continue;
        p_current_class=probablity(dictonary,x,current_classes)    
        if(first_run or p_current_class>best_p):
            best_p=p_current_class
            best_class=current_classes
        first_run=False
    return best_class    
```

```python
# we are going to predict the class for the test data
def predict(dictonary,x_test):
    y_pred=[]
    for x in x_test:
        x_class=predictsinglepoint(dictonary,x)
        y_pred.append(x_class)
    return y_pred
```
